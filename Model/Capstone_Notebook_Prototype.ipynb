## Data Preparation

```python
# Function to clean whitespaces in columns values
def clean_data(x):
    if isinstance(x, list):
        return [str.lower(i.replace(" ", "")) for i in x]
    else:
        #Check if director exists. If not, return empty string
        if isinstance(x, str):
            return str.lower(x.replace(" ", ""))
        else:
            return ''
```

```python
# Function to soup all the features
def create_soup(x):
    return x['Kategori'] + ' ' + str(x['Kebutuhan Pembiayaan Kredit (Juta)']) + ' ' + str(x['Jumlah Tenaga Kerja']) + ' ' + x['Provinsi']
```

```python
# Set the features to merge
features = ['Kategori', 'Kebutuhan Pembiayaan Kredit (Juta)', 'Provinsi']
```

```python
# Applying the clean_data function to features
df_train = df.copy()
for feature in features:
    df_train[feature] = df_train[feature].apply(clean_data)
```

```python
df_train.head(5)
```

```python
features_used = ['Nama','Kategori', 'Kebutuhan Pembiayaan Kredit (Juta)', 'Jumlah Tenaga Kerja', 'Provinsi']
for column in df_train:
  if column not in features_used:
    df_train = df_train.drop(column,axis=1)
    
```

#Getting User Input to calculate similarity

```python
# Getting User Profile as Input
data = {'Nama':['Investor 1'],
         'Kategori':['Industri Pengolahan'],
         'Provinsi':['DKI Jakarta'],
         'Jumlah Tenaga Kerja':[5.0],
         'Kebutuhan Pembiayaan Kredit (Juta)':[15.0]}
user_df = pd.DataFrame(data)
user_df
```

```python
# Appending the user data to item dataframes
df_soup = pd.concat([df_train, user_df]).reset_index()
df_soup
```

```python


# Applying the soup function to features
df_soup['soup'] = df_train.apply(create_soup, axis=1)
df_soup['soup']=df_soup['soup'].apply(str)
```

```python
df_soup[['soup']].head(2)
```

# Calculating the similarity

```python
#calculate the similarity

count = CountVectorizer()
count_matrix = count.fit_transform(df_soup['soup'])
cosine_sim2 = cosine_similarity(count_matrix, count_matrix)
```

```python
#Getting the index
indices = pd.Series(df_soup.index, index=df_soup['Nama'])
```

#Getting the 10 most matched with user

```python
def get_recommendations(user_name, cosine_sim):
    # Get the index of the user_name
    idx = indices[user_name]

    # Get the pairwsie similarity scores
    sim_scores = list(enumerate(cosine_sim[idx]))

    # Sort based on the similarity scores
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)

    # Get the scores of the 10 most match items with user
    sim_scores = sim_scores[1:11]

    # Get the item indices
    umkm_indices = [i[0] for i in sim_scores]

    # Return the top 10 most similar items
    return df.iloc[umkm_indices, :]
```

```python
result = get_recommendations('Investor 1', cosine_sim2)
```
